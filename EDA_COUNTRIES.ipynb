```python
axes,fig=plt.subplots(0,1,figsize=(15,5))
sns.scatterplot('country',"lifeExpectancy",hue="year",data=df2)
plt.xticks(rotation=90)
plt.show()
```

**Angola in the African continent is a place where the average age of people is still between 45-50 and the reason for that are poverty, civil violence, food shortages, political unrest, and lack of adequate medical services**

**Zimbabwe was showing a positive growth in life expectancy but had a fall during the 2000 due to the political unrest that happened there**


```python
df3=df[df["continent"]=="Europe"]
```

```python
axes,fig=plt.subplots(0,1,figsize=(15,10))
sns.scatterplot('country',"lifeExpectancy",hue="year",data=df3)
plt.xticks(rotation=90)
plt.show()
```

**Turkey in Europe has shown less growth in life expectancy as compared to the other countries in Europe**

**Here again it can be noticed that turkey has less gdp growth as compared to other and this may be the reason of low expectancy rate in Turkey.**

```python
axes,fig=plt.subplots(0,1,figsize=(15,10))
sns.scatterplot('country',"population",hue="year",data=df3)
plt.xticks(rotation=90)
plt.show()
```

**Germany and france has huge population in the Europe continent**

**Turkey has a huge population with low gdp resulting in least life expectancy in the Europe continent.**
